```python
def percentile_missing(df):
    percent_nan = 100 * df.isnull().sum()/ len(df)
    percent_nan = percent_nan[percent_nan > 0].sort_values()
    
    return percent_nan
```

QUESTION 1: The label we are trying to predict is the SalePrice column. Separate out the data into X features and y labels

```python
X = final_df.drop('SalePrice', axis = 1)
```

```python
y = final_df['SalePrice']
```

QUESTION 2: Use scikit learn to split up X and y into a training set and test set. Since we will later be using a Gridsearch stratergy, set your test proportion to 10%. To get the same data split as the solutions notebook, you can specify random_state = 101

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=101)
```

QUESTION 3: The dataset features has a variety of scales and units. For optimal regression performance, scale the X features, Take carefully note of what to use for fit() vs what to use for transform()

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

scaler.fit(X_train)
```

```python
X_train = scaler.transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

QUESTION 4: We will use an Elatic Net model. Create an instance of default ElasticNet model with scikit learn and also create a GridsearchCV object and search for the best parameters for your model based on your scaled training data.

```python
from sklearn.linear_model import ElasticNet
```

```python
base_elastic_model = ElasticNet(max_iter = 1000000)
```

```python
param_grid = {'alpha' : [0.1, 1, 5, 10, 50, 100],
             'l1_ratio' : [0.1, 0.3, 0.5, 0.7, 0.93, 0.98, 1]}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid_model = GridSearchCV(estimator = base_elastic_model, 
                         param_grid = param_grid, scoring = 'neg_mean_squared_error',
                         cv = 10, verbose = 1)
```

```python
grid_model.fit(X_train, y_train)
```

QUESTION 5: Display the best combinations of parameters for your model

```python
grid_model.best_params_
```

QUESTION 6: Evaluvate your models performance on the unseen 10% scaled testset.

```python
y_pred = grid_model.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
mean_absolute_error(y_test, y_pred)
```

```python
np.sqrt(mean_squared_error(y_test, y_pred))
```
