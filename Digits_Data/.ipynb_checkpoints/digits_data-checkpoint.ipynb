```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('33912190-digits.csv')
df.head()
```

```python
pixels = df.drop('number_label', axis = 1)
```

```python
pixels
```

```python
single_image = pixels.iloc[0]
```

```python
single_image
```

```python
single_image.to_numpy().shape
```

```python
number = single_image.to_numpy().reshape(8, 8)
```

```python
number
```

```python
plt.imshow(number, cmap = 'gray')
```

```python
sns.heatmap(number, annot = True)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaled_pixels = scaler.fit_transform(pixels)
```

```python
scaled_pixels
```

```python
from sklearn.decomposition import PCA
```

```python
pca_model = PCA(n_components=2)
```

```python
pca_pixels = pca_model.fit_transform(scaled_pixels)
```

```python
pca_model.explained_variance_ratio_
```

```python
np.sum(pca_model.explained_variance_ratio_)
```

```python
pca_pixels
```

```python
pca_pixels[:, 1]
```

```python
plt.figure(figsize=(10, 6), dpi = 150)
sns.scatterplot(pca_pixels[:, 0], pca_pixels[:, 1], hue = df['number_label'].values, palette = 'Set1')
plt.legend(loc = (1.05, 0))
```

## 3D Plot ##

```python
pca_model = PCA(n_components=3)
```

```python
pca_pixels = pca_model.fit_transform(scaled_pixels)
```

```python
from mpl_toolkits import mplot3d
```

```python
# %matplotlib notebook


plt.figure(figsize=(8, 8), dpi = 150)
ax = plt.axes(projection = '3d')
ax.scatter3D(pca_pixels[:, 0], pca_pixels[:, 1], pca_pixels[:, 2], c = df['number_label']);
```
